```python
dfp.to_parquet('people.parquet')
```

```python
table = pa.Table.from_pandas(dfp)
print(table)
print("Saving to parquet files...")
```

```python
table = pq.read_table('people.parquet',columns=['Index','First Name','Last Name','Sex','Job Title'])
table.to_pandas()
```

```python
table
```

```python
# write to multiple partitioned files
pq.write_to_dataset(table, root_path='my_parq2', partition_cols=['Job Title'])
```

```python
# read partitioned table
table = pq.read_table('my_parq2')
table.to_pandas().head()
```

```python
table.to_pandas()
```

```python
import pyarrow.compute as pc
grouped_table = table.group_by("Job Title").aggregate(
  [("Index",
  "count",  #count #max #min

  pc.CountOptions(mode="only_valid"))]
)

grouped_table.to_pandas()
```
